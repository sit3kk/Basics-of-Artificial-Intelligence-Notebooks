# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mean = [0, 0]
cov = [[1, 0], [0, 1]]  


x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

rv = multivariate_normal(mean, cov)

plt.figure(figsize=(8, 6))
plt.contourf(x, y, rv.pdf(pos), levels=100, cmap='viridis')
plt.colorbar()
plt.title('Density of a Two-dimensional Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
plt.figure(figsize=(8, 6))
plt.contour(x, y, rv.pdf(pos), colors='black')
plt.title('Contour Lines of a Two-dimensional Normal Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```

## Narysujmy poziomice (contour plot) dla mieszanki gaussów

Rozpatrzmy mieszankę gaussów postaci
$$
\frac{1}{2}N(m_1,\Sigma_1)+\frac{1}{2}N(m_2,\Sigma_2).
$$
Proszę narysować poziomice dla przykładu gdy $m_1\neq m_2$, $\Sigma_1=\Sigma_2$

```python

mean1 = [0, -1]  
mean2 = [0, 1]  
covariance = [[1, 0], [0, 1]] 


rv1 = multivariate_normal(mean1, covariance)
rv2 = multivariate_normal(mean2, covariance)


mixture_density = 0.5 * rv1.pdf(pos) + 0.5 * rv2.pdf(pos)


plt.figure(figsize=(8, 6))
plt.contourf(x, y, mixture_density, levels=100, cmap='viridis')
plt.colorbar()
plt.title('Contour Plot for a Mixture of Gaussians')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

Proszę wylosować na dwa sposoby: korzystając z scipy.stats.multivariate_normal oraz numpy.random.multivariate_normal

```python
from scipy.stats import multivariate_normal
import numpy as np


mean = [0, 0]
cov = [[1, 0], [0, 1]]


sample_scipy = multivariate_normal.rvs(mean, cov, size=1000)


sample_numpy = np.random.multivariate_normal(mean, cov, 1000)


plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)
plt.plot(sample_scipy[:, 0], sample_scipy[:, 1], 'o', markersize=3)
plt.title('Samples using scipy.stats.multivariate_normal')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')


plt.subplot(1, 2, 2)
plt.plot(sample_numpy[:, 0], sample_numpy[:, 1], 'o', markersize=3)
plt.title('Samples using numpy.random.multivariate_normal')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')

plt.tight_layout()
plt.show()
```
